```python
class Task:
  def __init__(self, name, arrival_time, completion_time, deadline, priority):
    self.name = name
    self.arrival_time = arrival_time
    self.completion_time = completion_time
    self.deadline = deadline
    self.priority = priority
    self.child = []
    self.parent = None

  def __str__(self):
    return self.name

  def add_child(self, child):
    self.child.append(child)

  def set_parent(self, parent):
    self.parent = parent

  def get_name(self):
    return self.name

  @staticmethod
  def print_tasks(tasks):
    for index, task in enumerate(tasks):
      print(tasks[index])
    print("\n")

  @staticmethod
  def sort_tasks(tasks):
    tasks.sort(key = lambda Task: Task.priority)

  @staticmethod
  def calculate_finish_time(current_time, arrival_time, completion_time):
    if current_time <= arrival_time:
      finish_time = arrival_time + completion_time
    else:
      finish_time = current_time + completion_time
    return finish_time
```

```python
from collections import deque

def bratleys_algorithm(tasks):
    tasks.sort(key=lambda t: (t.deadline, -t.priority))  # Sort tasks by deadline, then by priority (descending)
    current_time = 0

    for task in tasks:
        current_time = Task.calculate_finish_time(current_time, task.arrival_time, task.completion_time)
        task.finish_time = current_time
        task.lateness = task.finish_time - task.deadline

    return tasks

def print_schedule(tasks):
    print("Task\tArrival\tCompletion\tDeadline\tPriority\tFinish\tLateness")
    for task in tasks:
        print(f"{task.name}\t{task.arrival_time}\t{task.completion_time}\t\t{task.deadline}\t\t{task.priority}\t\t{task.finish_time}\t{task.lateness}")

tasks = [
    Task("task1", 4, 2, 7, 1),
    Task("task2", 1, 1, 5, 2),
    Task("task3", 1, 2, 6, 3),
    Task("task4", 0, 2, 4, 4),
]

scheduled_tasks = bratleys_algorithm(tasks)
print_schedule(scheduled_tasks)
```
